```python
import re
import numpy as np
from gensim.models import Word2Vec
from sklearn.cluster import KMeans
```

```python
with open("text_data.txt","r") as f:
    text = f.read()

print(text)
```

```python
sentences = re.split(r'[.!?]',text)
sentences = [s.strip() for s in sentences if s.strip()]
print(sentences)
```

```python

stopwords = ['a','an','the','is','of','that','on','with','in','and','more','as']
processed_sentences = []
for sent in sentences:
    sent = sent.lower()
    sent = re.sub(r'[^a-z\s]','',sent)
    words = sent.split()
    words = [w for w in words if w not in stopwords]
    processed_sentences.append(words)

print(processed_sentences)
```

```python
#word embeddings
model = Word2Vec(
    processed_sentences,
    vector_size = 100,
    window = 5,
    min_count = 1,
    sg = 1
)
```

```python
#sentence vectors
sentence_vectors = []
for words in processed_sentences:
    vector = []
    for w in words:
        if w in model.wv: 
            vector.append(model.wv[w])
    
    if len(vector) == 0:
        sentence_vectors.append(np.zeros(model.vector_size))
    else:
        sentence_vectors.append(np.mean(vector,axis = 0))

sentence_vectors = np.array(sentence_vectors)
```

```python
k = 2
kmeans = KMeans(n_clusters = k,random_state = 0)
kmeans.fit(sentence_vectors)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print(labels)
```

```python
summary = []

for i in range(k):
    cluster_indices = np.where(labels == i)[0]
    cluster_vectors = sentence_vectors[cluster_indices]
    centroid = centroids[i]
    distances = np.linalg.norm(cluster_vectors - centroid,axis = 1)
    closest_index = cluster_indices[np.argmin(distances)]
    summary.append(sentences[closest_index])

summary = '. '.join(summary)
print(summary)
```
