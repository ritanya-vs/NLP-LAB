Using Built-In TF-IDF

```python
import re
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
with open("text_data.txt","r") as f:
    text = f.read()

print(text)
```

```python
sentences = re.split(r'[.!?]',text)
sentences = [s.strip() for s in sentences if s.strip()]
print(sentences)
```

```python
#sentence vectors
vectorizer = TfidfVectorizer(stop_words = 'english')
tfidf_matrix = vectorizer.fit_transform(sentences)
tfidf_array = tfidf_matrix.toarray()
print(tfidf_array)
```

```python
sentence_scores = []
for row in tfidf_array:
    tfidf_sum = np.sum(row)
    n = np.count_nonzero(row)
    if n == 0:
        score = 0
    else:
        sentence_scores.append(tfidf_sum/n)

print(sentence_scores)
```

```python
ranked = []
for i in range(len(sentences)):
    ranked.append((sentence_scores[i],sentences[i]))

ranked = sorted(ranked,reverse = True)
print(ranked)
```

```python
top_n = 2
result = []
for i in range(top_n):
    result.append(ranked[i][1])

summary = '. '.join(result)
print(summary)
```

Manual Code - Tf-Idf

```python
import re
import math
from collections import Counter
```

```python
sentences = re.split(r'[.!?]',text)
sentences = [s.strip() for s in sentences if s.strip()]
print(sentences)
```

```python
processed_sentences = []
stopwords = ['a','an','the','is','of','that','on','with','in','and','more','as']

for sent in sentences:
    sent = sent.lower()
    sent = re.sub(r'[^a-z\s]','',sent)
    words = sent.split()
    words = [w for w in words if w not in stopwords]
    processed_sentences.append(words)

print(processed_sentences)
```

```python
vocab = set()
for sent in processed_sentences:
    vocab.update(sent)

vocab = list(vocab)
print(vocab)
```

```python
#Term Frequency
tf = []
for sent in processed_sentences:
    word_count = Counter(sent)
    sent_tf = []
    for word in vocab:
        sent_tf.append(word_count[word]/len(sent))
    tf.append(sent_tf)

tf = np.array(tf)
```

```python
#Document Frequency
df = {}
for word in vocab:
    count = 0
    for sent in processed_sentences:
        if word in sent:
            count += 1
    df[word] = count

#Inverse Document Frequency
idf = []
N = len(processed_sentences)
for word in vocab:
    idf.append(math.log(N/(df[word] + 1)))

idf = np.array(idf)
```

```python
tfidf_matrix = tf*idf
print(tfidf_matrix)
```

```python
sentence_scores = []
for row in tfidf_matrix:
    tfidf_sum = np.sum(row)
    n = np.count_nonzero(row)
    if n == 0:
        score = 0
    else:
        sentence_scores.append(tfidf_sum/n)

print(sentence_scores)
```

```python
ranked = []
for i in range(len(sentences)):
    ranked.append((sentence_scores[i],sentences[i]))

ranked = sorted(ranked,reverse = True)
print(ranked)
```

```python
top_n = 2
result = []
for i in range(top_n):
    result.append(ranked[i][1])

summary = '. '.join(result)
print(summary)
```
